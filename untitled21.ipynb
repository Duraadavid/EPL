```python
import pandas as pd

# Creating a DataFrame from the transcribed data
data = {
    "Home": ["MCI", "MCI", "MCI"],
    "Away": ["MNU", "BOU", "BRI"],
    "Home_Goals_1": [1, 1, 2],
    "Away_Goals_1": [0, 1, 0],
    "Home_Goals_2": [2, 3, 2],
    "Away_Goals_2": [1, 0, 1],
    "Home_Goals_3": [2, 3, 1],
    "Away_Goals_3": [2, 1, 0],
    "Home_Goals_4": [1, 1, 1],
    "Away_Goals_4": [0, 1, 1],
    "Home_Goals_5": [1, 2, 1],
    "Away_Goals_5": [2, 0, 2],
    "Home_Goals_6": [1, 2, 2],
    "Away_Goals_6": [0, 3, 1]
}

df = pd.DataFrame(data)
df
```

```python
# Melt the DataFrame to have one row per match occurrence
df_melted = df.melt(id_vars=["Home", "Away"],
                    value_vars=["Home_Goals_1", "Home_Goals_2", "Home_Goals_3", "Home_Goals_4", "Home_Goals_5", "Home_Goals_6"],
                    var_name="Occurrence", value_name="Home_Goals")

# Extract Away Goals corresponding to each Home Goals
away_goals = df.melt(id_vars=["Home", "Away"],
                     value_vars=["Away_Goals_1", "Away_Goals_2", "Away_Goals_3", "Away_Goals_4", "Away_Goals_5", "Away_Goals_6"],
                     var_name="Occurrence", value_name="Away_Goals")["Away_Goals"]

df_melted["Away_Goals"] = away_goals

# Calculate total goals and filter by conditions
df_melted["Total_Goals"] = df_melted["Home_Goals"] + df_melted["Away_Goals"]
df_filtered = df_melted[(df_melted["Total_Goals"] > 2) & (df_melted["Total_Goals"] < 4)]

# Create labels for home win/draw
df_filtered["Home_Win_Draw"] = (df_filtered["Home_Goals"] >= df_filtered["Away_Goals"]).astype(int)

df_filtered
```
