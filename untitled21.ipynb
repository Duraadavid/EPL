```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import mean_squared_error, accuracy_score

# Load the data
data = pd.read_csv('//content/10% DATA BET - Sheet1 (1).csv')

# Data inspection
data.head()
```

```python
import pandas as pd
# Melt the DataFrame to have one row per match occurrence
melted_data = pd.melt(data, id_vars=['Home', 'Away'],
                       value_vars=[col for col in data.columns if 'Home Goals' in col],
                       var_name='Match',
                       value_name='HomeGoalsValue') # Changed 'Home Goals' to 'HomeGoalsValue'

# Extract Away Goals corresponding to each Home Goals
melted_data['Away Goals'] = melted_data.apply(lambda row: data.loc[(data['Home'] == row['Home']) & (data['Away'] == row['Away']), row['Match'].replace('Home Goals', 'Away Goals ')].values[0], axis=1)

# Calculate total goals
melted_data['Total Goals'] = melted_data['HomeGoalsValue'] + melted_data['Away Goals'] # Use the new column name

# Filter by conditions
filtered_data = melted_data[(melted_data['Total Goals'] > 0) & (melted_data['HomeGoalsValue'] != melted_data['Away Goals'])] # Use the new column name
filtered_data
```

```python
# prompt: Using dataframe filtered_data: # Extract Away Goals corresponding to each Home Goals

# Group the DataFrame by 'HomeGoalsValue' and aggregate 'Away Goals' using a list
result = filtered_data.groupby('HomeGoalsValue')['Away Goals'].apply(list)

# Print the result showing Away Goals corresponding to each Home Goals
print(result)
```

```python
# prompt: Using dataframe filtered_data: # Calculate total goals and filter by conditions

# Calculate the total goals for each match
filtered_data["Total Goals"] = filtered_data["HomeGoalsValue"] + filtered_data["Away Goals"]

# Filter the DataFrame for matches where "Home Goals" were greater than 2 and "Away Goals" were less than 1
filtered_data[(filtered_data["HomeGoalsValue"] > 2) & (filtered_data["Away Goals"] < 1)]
```

```python
# Melt the DataFrame to have one row per match occurrence
df_melted = data.melt(id_vars=["Home", "Away"],
                    value_vars=["Home Goals.1", "Away Goals.1", "Home Goals.2", "Away Goals.2", "Home Goals.3", "Home Goals.4", "Home Goals.5", "Home Goals.6"], # Correct column names
                    var_name="Occurrence", value_name="Home Goals")

# Extract Away Goals corresponding to each Home Goals
away_goals = data.melt(id_vars=["Home", "Away"],
                     value_vars=["Away Goals.1", "Away Goals.2", "Away Goals.3", "Away Goals.4", "Away Goals.5", "Away Goals.6"], # Correct column names
                     var_name="Occurrence", value_name="Away Goals")["Away Goals"]

df_melted["Away Goals"] = away_goals

# Calculate total goals and filter by conditions
df_melted["Total Goals"] = df_melted["Home Goals"] + df_melted["Away Goals"]
df_filtered = df_melted[(df_melted["Total_Goals"] > 2) & (df_melted["Total_Goals"] < 4)]

# Create labels for home win/draw
df_filtered["Home_Win_Draw"] = (df_filtered["Home Goals"] >= df_filtered["Away Goals"]).astype(int)

df_filtered
```
